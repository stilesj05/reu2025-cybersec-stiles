```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline

model_name = "CyberNative/CyberBase-13b"

bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True
)

# (1) Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# (2) Define a very minimal device_map, letting HF decide the rest:
#     - “sequential” will fill up your GPU with layers until it runs out of room,
#       then offload the remaining layers to CPU **in fp32** (because of llm_int8_enable_fp32_cpu_offload=True).
device_map = "sequential"

# (3) Load the model with quantization + CPU offload. ‘device_map="sequential"’
#     automatically assigns as many 8-bit layers as will fit on GPU. The rest live on CPU.
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map=device_map,
    trust_remote_code=True
)

# (4) Create your text-generation pipeline
generator = pipeline("text-generation", model=model, tokenizer=tokenizer)
```

```python
output = generator("Explain how ransomware works and how to defend against it.", max_length=200)
print(output[0]["generated_text"])
#Correct output for Part 2
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, pipeline

# ———————————————————————————————
# Configuration for Foundation-Sec-8B
# ———————————————————————————————

model_name3 = "fdtn-ai/Foundation-Sec-8B"

# 8-bit + CPU offload config
bnb_config3 = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True
)

# (1) Tokenizer (add `use_auth_token=True` if the repo is gated)
tokenizer3 = AutoTokenizer.from_pretrained(
    model_name3,
    use_auth_token=True
)

# (2) Model (8-bit quant + pack‐on‐GPU then spill to CPU)
model3 = AutoModelForCausalLM.from_pretrained(
    model_name3,
    quantization_config=bnb_config3,
    device_map="sequential",   # “fill GPU, spill rest (in fp32) to CPU”
    trust_remote_code=True,
    use_auth_token=True
)

# (3) Pipeline
generator3 = pipeline("text-generation", model=model3, tokenizer=tokenizer3)

```

```python
prompt = (
    "Map CVE-2023-24055 to its related CWE and explain the underlying weakness, "
    "then provide mitigation suggestions."
)
output = generator3(prompt, max_new_tokens=500)
print(output[0]["generated_text"])
#Correct output for Foundation-Sec-8B
```
